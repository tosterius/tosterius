```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.00001, 0.9999991, 0.002)
y = 1 / 2 * np.log((1 - x) / x)
```

```python
plt.plot(x, y, color='r')
plt.grid(True, which='both')
plt.xlabel('misclassification error')
plt.ylabel('weight of hypothesis')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
```

```python
x = np.arange(0, 4, 0.2)
plt.plot(x, np.exp(x), color='b')

plt.grid(True, which='both')
plt.xlabel('weight of hypothesis')
plt.ylabel('exp(+)')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
```

```python
x = np.arange(0, 4, 0.2)
plt.plot(x, np.exp(-x), color='b')

plt.grid(True, which='both')
plt.xlabel('weight of hypothesis')
plt.ylabel('exp(-)')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
```
